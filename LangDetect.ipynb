```python
data = pd.read_csv('/content/drive/MyDrive/Thesis/Data/motorcycle.csv')
data.head()
```

```python
Bangla = []
English = []
Romanized = []
garbage = []

for i,v in data.iterrows():
  text = str(v['selection1_Reviews_name'])
  try:
    lang = detect(text)
    if lang =='en':
      English.append(text)
    elif lang =='bn':
      Bangla.append(text)
    else:
      Romanized.append(text)
  except:
    garbage.append(text)

print(Bangla)
print('\n')
print(English)
print('\n')
print(Romanized)
  
```

```python
banglaData = pd.DataFrame({'Bangla Reviews': Bangla})
englishData = pd.DataFrame({'English Reviews': English})
romanizedData = pd.DataFrame({'Romanized Reviews': Romanized})

banglaData = banglaData.drop_duplicates(subset="Bangla Reviews")
englishData = englishData.drop_duplicates(subset="English Reviews")
romanizedData = romanizedData.drop_duplicates(subset="Romanized Reviews")




finalBangla = pd.read_excel('/content/drive/MyDrive/Thesis/Data/FInalized Data/New Bangla .xlsx')
finalEnglish = pd.read_excel('/content/drive/MyDrive/Thesis/Data/FInalized Data/New English.xlsx')
finalRomanized = pd.read_excel('/content/drive/MyDrive/Thesis/Data/FInalized Data/New Romanized.xlsx')

framesEnglish = [finalEnglish, englishData]
newEnglish = pd.concat(framesEnglish)
newEnglish = newEnglish.drop_duplicates(subset='English Reviews')

framesBangla = [finalBangla, banglaData]
newBangla = pd.concat(framesBangla)
newBangla = newBangla.drop_duplicates(subset='Bangla Reviews')

framesRomanized = [finalRomanized, romanizedData]
newRomanized = pd.concat(framesRomanized)
newRomanized = newRomanized.drop_duplicates(subset='Romanized Reviews')

fileEnglish = "/content/drive/MyDrive/Thesis/Data/FInalized Data/New English.xlsx"
newEnglish.to_excel(fileEnglish, index=False)

fileBangla = "/content/drive/MyDrive/Thesis/Data/FInalized Data/New Bangla .xlsx"
newBangla.to_excel(fileBangla, index=False)

fileRomanized = "/content/drive/MyDrive/Thesis/Data/FInalized Data/New Romanized.xlsx"
newRomanized.to_excel(fileRomanized, index=False)

print('Size of Final English Dataset: ', len(finalEnglish))
print('Size of new English Dataset: ', len(englishData))
print('Size After concatanation: ', len(newEnglish))

print('\n\n')

print('Size of Final Bangla Dataset: ', len(finalBangla))
print('Size of new Bangla Dataset: ', len(banglaData))
print('Size After concatanation: ', len(newBangla))

print('\n\n')

print('Size of Final Romanized Dataset: ', len(finalRomanized))
print('Size of new Romanized Dataset: ', len(romanizedData))
print('Size After concatanation: ', len(newRomanized))
```
