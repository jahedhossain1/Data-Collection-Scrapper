```python
 headers = {"Accept-Language": "utf ;q=0.5"}
 
pages = [str(i) for i in range(1, 101)]
#pages
```

```python
from time import sleep, time
from random import randint
import requests
from bs4 import BeautifulSoup
 
names = []
dates = []
stars = []
headlines = []
bodies = []
 
URL = "https://www.trustpilot.com/review/www.daraz.com.bd?page="
start_time = time()
for i in pages[:10]:
  url = URL+i
  response = requests.get(url, headers= headers)
  sleep(randint(1, 3))
  total_time = time() - start_time
 
  soup = BeautifulSoup(response.content, 'html.parser')
  containers = soup.find_all('div', class_="review-card")
  x = 0
  for con in containers:
    if con is not None:
      #name = con.find('div', class_='consumer-information__name').text.strip().title()
      review_body = con.find('div', class_="review-content__body")
      review_header = con.find('div', class_="review-content-header")
      title = review_body.h2
      body = review_body.p
      star = review_header.find('div', class_='star-rating star-rating--medium').img['alt'].strip()[0]
      date = review_header.find('div', class_='review-content-header__dates').text.split(',')[0].strip().split(":")[1].split("T")[0].split('"')[1]
      if None not in (name, review_body, review_header, body, title, star, date):
        the_name = name
        names.append(the_name)
        headline = title.text.strip()
        headlines.append(headline)
        the_body = body.text.strip()
        bodies.append(the_body)
        the_date = date
        dates.append(the_date)
        stars.append(star)
```

```python
import pandas as pd

df = pd.DataFrame({
    'Star': stars,
    'Title': headlines,
    'Body': bodies
})
df
```
